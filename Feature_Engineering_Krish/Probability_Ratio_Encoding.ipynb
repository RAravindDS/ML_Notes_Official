```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('train_titanic.csv', usecols = ['Cabin','Survived'])
```

```python
data.head()
```

```python
data['Cabin'].fillna('Missing', inplace = True)
```

```python
data['Cabin'].unique()
```

```python
data['Cabin'] = data['Cabin'].astype(str).str[0]
```

```python
data['Cabin'].unique()
```

```python
prob = data.groupby(['Cabin'])['Survived'].mean() # finding the percentage.
```

```python
prob = pd.DataFrame(prob)
```

```python
prob.head()
```

```python
prob['died'] = 1-prob['Survived']
```

```python
prob.head()
```

```python
prob['ratio'] = prob['Survived']/prob['died']
```

```python
prob
```

```python
a = prob['ratio'].to_dict()
```

```python
data['Cabin'] = data['Cabin'].map(a)
```

```python
data.head()
```

```python
data.drop('died',axis =1, inplace = True)
```

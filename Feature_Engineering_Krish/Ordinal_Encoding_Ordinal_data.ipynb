* if we have more number of categorical data in the data set for ordinal dtype, it tough to do the one hot encoding, so we use the ordinal encoding or label encoding here
* we can use for fixed number of categories, so its too easier.

```python
import pandas as pd
import numpy as np
import datetime
```

```python
base = datetime.datetime.today()
base_list = [ base - datetime.timedelta(days = x) for x in range(0,20)]
df = pd.DataFrame(base_list)
df.columns = ['day']
```

```python
df
```

```python
li = ['monday','tuesday','wednesday','thursday','friday','saturday','sunday'
     'monday','tuesday','wednesday','thursday','friday','saturday','sunday',
     'monday','tuesday','wednesday','thursday','friday','saturday','sunday']
```

```python
df['dat_of_week'] = li
```

```python
df
```

```python
print(li)
```

```python
dic = {'monday':1,'tuesday':2,'wednesday':3,'thursday':4,'friday':5,'saturday':6,'sunday':7}
```

```python
df['dat_of_week'] = df['dat_of_week'].map(dic)
```

```python
df['dat_of_week'] = df['dat_of_week'].values.astype(int)
```

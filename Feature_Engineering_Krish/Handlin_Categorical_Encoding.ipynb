### 3. Count of Frequency encoding

```python
train_set = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data' , 
                        header = None,index_col=None) 
train_set.head()
```

```python
col = [1,3,5,6,7,8,9,13]
```

```python
train = train_set[col]
```

```python
train.columns = ['Employment', 'Degree', 'Status', 'Designation', 'family_job', 'Race', 'Sex','Country']
```

```python
a = train['Country'].value_counts().to_dict()
```

```python
train['Country'] = train['Country'].map(a)
```

```python
train.head()
```

### 4. Mean Encoding

```python
data = pd.read_csv('train_titanic.csv', usecols = ['Cabin','Survived'])
```

```python
data.head()
```

```python
data.Cabin.fillna('theriyathu', inplace = True)
```

```python
di = data.groupby(['Cabin'])['Survived'].mean().to_dict() # just checking the nr of percentage pepole survived
```

```python
data.Cabin = data.Cabin.map(di)
```

```python
data.head()
```

### 2. Target Guided Ordianl Encoding

```python
data = pd.read_csv('train_titanic.csv', usecols = ['Cabin','Survived'])
```

```python
data.head(15)
```

```python
data.Cabin.fillna('theriyathu', inplace = True)
```

```python
data['Cabin'] = data.Cabin.astype(str).str[0]
```

```python
data.groupby(['Cabin'])['Survived'].mean() # just checking the nr of percentage pepole survived
```

```python
lab = ['T', 't', 'A', 'G', 'C', 'F', 'B', 'E', 'D']
```

```python
lab2 = {k:i for i,k in enumerate(lab, 0) }
```

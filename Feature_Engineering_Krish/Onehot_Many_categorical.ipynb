* We're going to take first 10 to 20 most frequent features and we do the encoding remaining all the features are 0, tooked features are 1.

```python
data.X2.value_counts().head(20)
# This for first features, we select the first 10 and do the label encoding and remaining all are zero.
```

```python
# let's take the top 10 values to do the encoding

top_10 = [i for i in data.X2.value_counts().sort_values(ascending = True).head(10).index]
```

```python
top_10 # top 10 features this values are all 1 and remaining all are zero.
```

```python
for i in top_10:
    data[i] = np.where(data['X2'] == i, 1, 0)
```

```python
data[['X2'] + top_10].head()
```

```python
# I am going to take smaller dataset
data = data.iloc[:,2:7]
```

```python
data.head()
```

```python
# I am going to do encoding fo all
```

```python
def one_hot(data, variable, top):
    for i in top:
        data[variable +'_'+ i] = np.where(data[variable] == i, 1, 0)
```

```python
one_hot(data, 'X2', top_10)
```

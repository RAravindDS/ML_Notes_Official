## 1. Methods to Search for Matches

some matching methods are 

* match(): Determine if the RE matches at the beginning of the string.
* search(): Scan through a string, looking for any location where this RE matches.
* findall(): Find all substrings where the RE matches, and returns them as a list.
* finditer(): Find all substrings where the RE matches, and returns them as an iterator.

```python
test_string = '123abc456789abc123ABC'

""" Findall returns the string, and it will search everywhere """ 

pattern = re.compile(r'abc') 

matches = pattern.findall(test_string) 

a = [ i for i in matches ] 
a
```

```python
test_string = '123abc456789abc123ABC'

""" Match returns the string, that will apperar in the beginning of the string """ 

pattern = re.compile(r'abc') 

matches = pattern.match(test_string) 

print(matches) 

""" It gives none, because match look only begining of the strings """
```

```python
test_string = '123abc456789abc123ABC'

""" search returns the firstfinding string, and it will search everywhere """ 

pattern = re.compile(r'abc') 

matches = pattern.search(test_string) 

matches
```

```python
test_string = '123abc456789abc123ABC'

""" Finditer returns the string for iterator, and it will search everywhere """ 

pattern = re.compile(r'abc') 

matches = pattern.finditer(test_string) 

for match in matches: 
    print(match)

""" And after using find iter() we can do certain operations like group, start, end, span """
```

## 2. Methods on a Finditer match

```python
# match.span(), we will get the tuple here 
test_string = '123abc456789abc123ABC'

""" Finditer returns the string for iterator, and it will search everywhere """ 

pattern = re.compile(r'abc') 

matches = pattern.finditer(test_string) 
for match in matches: 
    print(match.span())

    
```

```python
# finditer()
my_string = 'abc123ABC123abc'
pattern = re.compile(r'123')
matches = pattern.finditer(my_string)
for match in matches:
    print(match)
    print(match.span(), match.start(), match.end()) # start and stop gives the index numbes
    print(match.group()) # It just group all 
```

## 3. Meta Characters

* Metacharacters are characters with a special meaning:
* All meta characters: **. ^ $ * + ? { } [ ] \ | ( )**

* . - We want to look for any character (except a new line character) 
* ^ - Starts with '^hello' 
* '$' - Ends with Words'$' (dollar Symbol look the metadata)
* '*' - '
* '+' - One or more occurence. 
* {} - Exactly specified number of characters. 
* [] - A set of characters [a-z] 
* | - Either or 'falls or stays' 
* () - Capture and group 
* \ - special sequence or escape characters '\d'
*** 
### More Metacharacters / Special Sequences
A special sequence is a \ followed by one of the characters in the list below, and has a special meaning:

* \d :Matches any decimal digit; this is equivalent to the class [0-9].
* \D : Matches any non-digit character; this is equivalent to the class [^0-9].
* \s : Matches any whitespace character;
* \S : Matches any non-whitespace character;
* \w : Matches any alphanumeric (word) character; this is equivalent to the class [a-zA-Z0-9_].
* \W : Matches any non-alphanumeric character; this is equivalent to the class [^a-zA-Z0-9_].
* \b Returns a match where the specified characters are at the beginning or at the end of a word r"\bain" r"ain\b"
* \B Returns a match where the specified characters are present, but NOT at the beginning (or at the end) of a word r"\Bain" r"ain\B"
* \A Returns a match if the specified characters are at the beginning of the string "\AThe"
* \Z Returns a match if the specified characters are at the end of the string "Spain\Z"



```python
# 1 '.' - Dot print very characther excpet new line 

test_string = '123abc456789abc123ABC'

pattern = re.compile(r'.') 

matches = pattern.finditer(test_string) 

for i in matches: 
    print(i.group())
```

```python
# \ this is the search characther

test_string = '123abc456789abc123ABC.'

pattern = re.compile(r'\.') 
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

```python
# ^ this ensures the starting charcters 

test_string = '123abc456789abc123ABC.'
pattern = re.compile(r'^123') 
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

```python
# $ this ensures the ending charcters 

test_string = '123abc456789abc123ABC'
pattern = re.compile(r'ABC$') 
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

## 5. Sets

A set is a set of characters inside a pair of square brackets [] with a special meaning. Append multiple conditions back-to back, e.g. [aA-Z].
A ^ (caret) inside a set negates the expression.
A - (dash) in a set specifies a range if it is in between, otherwise the dash itself.

Examples:
- [arn] Returns a match where one of the specified characters (a, r, or n) are present
- [a-n] Returns a match for any lower case character, alphabetically between a and n
- [^arn] Returns a match for any character EXCEPT a, r, and n
- [0123] Returns a match where any of the specified digits (0, 1, 2, or 3) are present
- [0-9] Returns a match for any digit between 0 and 9
- 0-5 Returns a match for any two-digit numbers from 00 and 59
- [a-zA-Z] Returns a match for any character alphabetically between a and z, lower case OR upper case

```python
# Matching multiple single character 

test_string = 'hello 123_AARV'

pattern = re.compile(r'[a-z0-9A-B]') # all the characters from a - z 
matches = pattern.finditer(test_string)
 
for i in matches: 
    print(i)
```

## 6. Quantifier: 

* '*' : 0 or more
* '+' : 1 or more
* ? : 0 or 1, used when a character can be optional
* {4} : exact number
* {4,6} : range numbers (min, max)

```python
test_string = 'hello_123' 

pattern = re.compile(r'\d*') # \d matches any decimal digits  
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

```python
test_string = 'hello123' 

pattern = re.compile(r'_?\d+') # \? optional and + 1 or more digit underscore it finds that optional
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

```python
test_string = 'hello123' 

pattern = re.compile(r'\d{1,2}') 
matches = pattern.finditer(test_string) 
for i in matches: 
    print(i)
```

## 8. Grouping

```python
emails = """
hello world
1223 
2020-05-20
Mr Simpson
Mrs Simpson
Mr. Brown
Ms Smith
Mr. T
pythonengineer@gmail.com
Python-engineer@gmx.de
python-engineer123@my-domain.org
"""
# I only want to extract the emails from the strings


pattern = re.compile(r'[a-zA-Z0-9-]+@[a-zA-Z-]+\.[a-zA-Z]+') # + 1 or more and it combines everything in a line 
pattern = re.compile(r'[a-zA-Z0-9-]+@[a-zA-Z-.]+')
pattern = re.compile(r'([a-zA-Z0-9-]+)@([a-zA-Z-]+)\.([a-zA-Z]+)') # we grouped over parenthesis 
matches = pattern.finditer(emails) 

for i in matches: 
    print(i.group(0))
    # print(i.group(1))
    # print(i.group(2))
    # print(i.group(3))
```

## 9. Modifications: 

There are two methods to modify the code 
* split(): Split the string into a list, splitting it wherever the RE matches
* sub(): Find all substrings where the RE matches, and replace them with a different string

```python
# split
my_string = '123abc456789abc123ABC'

pattern = re.compile(r'abc')
matches = pattern.finditer(my_string) 

splitted = pattern.split(my_string) 
print(splitted)

```

```python
# sub

test_string = 'hello world, you are the best wolrd' 

pattern = re.compile(r'world') 
subbed_string = pattern.sub('planet', test_string) 

print(subbed_string) # It just look the sentences and modifies into given word .... 
```

#### Small Example 2

```python
urls = """
http://python-engineer.com
https://www.python-engineer.org
http://www.pyeng.net
"""


pattern = re.compile(r'https?://(www\.)?([a-zA-Z-]+)(\.[a-zA-Z]+)')
matches = pattern.finditer(urls) 

for match in matches: 
    print(match)
    
subbed_urls = pattern.sub(r'\2\3', urls) # we just called group 2 and group 3 
print(subbed_urls)
```

## 10. Compilation flags: 

* ASCII, A : Makes several escapes like \w, \b, \s and \d match only on ASCII characters with the respective property.
* DOTALL, S : Make . match any character, including newlines.
* IGNORECASE, I : Do case-insensitive matches.
* LOCALE, L : Do a locale-aware match.
* MULTILINE, M : Multi-line matching, affecting ^ and $.
* VERBOSE, X (for ‘extended’) : Enable verbose REs, which can be organized more cleanly and understandably.

```python
my_string = "Hello World"
pattern = re.compile(r'world', re.IGNORECASE) # No match without I flag
matches = pattern.finditer(my_string)
for match in matches:
    print(match)

my_string = '''
hello
cool
Hello
'''
# line starts with ...
pattern = re.compile(r'^[a-z]', re.MULTILINE) # No match without M flag
matches = pattern.finditer(my_string)
for match in matches:
    print(match)
```

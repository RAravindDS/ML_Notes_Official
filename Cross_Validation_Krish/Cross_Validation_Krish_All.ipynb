```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = sns.load_dataset('titanic')
```

```python
data
```

```python
data.dropna(inplace = True)
```

```python
X = data[['pclass', 'age', 'sibsp', 'parch', 'fare']]
y = data['survived']# I am just taking for example
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf_fit = rf.fit(X_train, y_train)
rf_pred = rf_fit.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(y_test, rf_pred))
print(classification_report(y_test, rf_pred))
print(accuracy_score(y_test, rf_pred))
```

```python
# This was the normal method without any hyperparameter tunning and many things. 
# Let's too CV........
```

## K Fold 🎉

```python
from sklearn.model_selection import KFold
kfold = KFold(90)
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(rf, X, y, cv = kfold) 
```

### Stratified K fold Cross Validation.

```python
from sklearn.model_selection import StratifiedKFold

accuracy = []

skf = StratifiedKFold(n_splits = 100, random_state = None) # n_splits is nr of iteration.
skf.get_n_splits(X,y)


for train_index, test_index in skf.split(X, y): # The skf.split will give index of training and test data. 
    
    #print('Train', train_index, 'Test', test_index )
    X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
    y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]
    
    
    rf.fit(X1_train, y1_train)
    pred = rf.predict(X1_test)
    score = accuracy_score(pred, y1_test)
    accuracy.append(score)
    
print(accuracy)
```

```python
np.array(accuracy).mean() # We can increse the accuracy based on split and many things. 
```

## Or we can use this techinque also

```python
from sklearn.model_selection import StratifiedKFold

skfold = StratifiedKFold(n_splits = 100)
sco = cross_val_score(rf, X, y, cv = skfold)
print(np.mean(sco))
```

### Leave one out Cross Validation

```python
from sklearn.model_selection import LeaveOneOut
leave_out = LeaveOneOut()
s = cross_val_score(rf, X, y, cv = leave_out)
print(np.mean(s))
```

### Repeated Random CV

```python
from sklearn.model_selection import ShuffleSplit

splet = ShuffleSplit(n_splits = 10, test_size = 0.30)
res = cross_val_score(rf, X, y, cv = splet)
print(np.mean(res))
```

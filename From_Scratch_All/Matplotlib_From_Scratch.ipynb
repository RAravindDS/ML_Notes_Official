# Legend
* It will help you to find the accurate detalis on our data.

```python
b = np.array([1,2,3,4])
plt.plot(b**2,'r--',label= 'square')
plt.plot(b**3,'b-.', label = 'cube')
plt.legend(title = 'my legend') # we need to represent this
plt.xlabel(' x axid')
plt.ylabel(' y axis')
plt.grid(color = 'blue',alpha = 0.6)
```

# xticks and yticks
* it changes the direction of x and y axis representation.

```python
b = np.array([1,2,3,4])
plt.plot(b**2,'r--',label= 'square')
plt.plot(b**3,'b-.', label = 'cube')
plt.legend(title = 'my legend') # we need to represent this
plt.xlabel(' x axid')
plt.ylabel(' y axis')
plt.grid(color = 'blue',alpha = 0.6)
plt.xticks(rotation = 90,color = 'blue')
plt.yticks(rotation = 40,color = 'green');
plt.tick_params(length = 10, color = 'red')
```

# Subplots

```python
import pandas as pd

data = pd.read_csv('framingham.csv')
```

```python
data.head()
```

```python
educataion = data.education
cholorie = data.totChol
age = data.age
male = data.age
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 2)
axes[0,0].plot(educataion,'r-')
axes[0,1].plot(cholorie,'b-.')
axes[1,0].plot(age,'g*:')
axes[1,1].plot(male,'b^--')
```

```python
fig, axes = plt.subplots(2,2)
axes[0,0].plot(educataion,'r',label = 'summa')
axes[0,1].plot(male,'b',label = 'male')
axes[1,0].plot(age,'g',label = 'age')
axes[1,1].plot(cholorie, 'y', label ='cholorie')
axes[0,0].legend()
axes[0,1].legend()
axes[1,0].legend()
axes[1,1].legend()
axes[0,0].set_xlabel('summa x label',{'color':'green'})
axes[0,0].set_ylabel('summa y label',{'color':'blue'})
axes[0,1].set_xlabel(' x label',{'color':'green'})
axes[0,1].set_ylabel('y label',{'color':'blue'})
axes[1,0].set_xlabel('summa x label',{'color':'green'})
axes[1,0].set_xlabel('summa x label',{'color':'green'})
axes[1,1].set_ylabel('summa y label',{'color':'blue'})
axes[1,1].set_ylabel('summa y label',{'color':'blue'})
#t.xlabel(['summa','vamma','gamma','djkd'])
```

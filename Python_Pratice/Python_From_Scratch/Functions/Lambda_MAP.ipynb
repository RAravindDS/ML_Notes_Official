#### Lambda Function or Anonymous Function

* The Function which calls without a name that why this is annonymous function 

#### The Syntax is - lambda argument(one or more) : expression(it should contain only one expression)

```python
a = lambda x,y: x+y

print(a(1,2))
```

```python
b = lambda x,y: x-y
```

```python
print(b(8,9))
```

#### Map Function

* Used to apply function all the elements in sequence like if you want to apply function to all the numbers in list(withoutloop)

#### The syntax- map(function, sequence) 
* Sequence - list, typle, strings. 
* If you want to apply function each element of the sequence you can use this map function 

```python
a = [1,2,3,4,5,6] # I want squre for all

def square(a):
    return a*a

map(square,a) # map will return the iterator 
```

```python
list(map(square,a))
```

```python
d = ['1', '2', '3', '4', '5', '6'] # if you want to change str to int 

nu = list(map(lambda x: int(x),d))

type(nu[0])
```

```python
d = [2,3,24,2]
```

```python
nu = list(map(lambda x: x*x, d))
nu
```

```python
# Default 

def bio(name, age):
    print(name)
    print(age)
#bio(18)


def bio(name, age = 18): # default argument 
    print(name)
    print(age)
    
bio('aravind') # see it now workss. .. .. .. . .. .. .. .. .. 
```

### *varialbe length argument 

#### If you want to pass multiple actual arguments inside the function means we can use this


* It stores in typle
* Args should be come at end of the function otherwise it will throw an error

```python
# Numbers is not a argument.. 

def add(a,b):
    print(a,b)
    
    
#add(8,8,1,2,3,4)

# For that we need to use tuple to do that. ()

def add(a, *b): 
    print(a,b)
    
add(8,8,1,2,3,4,5) # now the remaining value will store as tuple to b

```

```python
# For accessing the tuple we can use the for loop to retrive the valuess. .. .. . . 

def add(a,*b):
    for i in b: 
        a = a+i 
    print(a)
    

add(8,8,1,2,3,4,5)
```

### **K Wargs ( keyword argument)

* It stores in dictionary 



```python
def person(name, *others):
    print(name, others)
    
#person('aravind')

#person('aravind', 'summa', 'pondy', 1111)

# ________________________________VARIALE LENGHT ARGUMENT___________________________________________________

# IF Want to Save the extra actual argument using keyword we can use this technique.. 

def person(name, **others): # It stores in dictionary
    print(name, others)
    return others 
    
a = person('aravind', age = 1, state = 'TamilNadu', phoneno= 9139787)
```

## Scope 

```python
a = 10  # Global Variables

def function():
    a = 14 # Local Variables 
    print(a)
    
function() # Local variable printed
print(a)   # Global variable printed 
```

```python
a = 10
def function():
    print(a)
    
function()
```


* Scope - Accessing inside the loop after it will search globally

Scope: 
- First it will look on local scope (inside the function) if not it will go to global scope

```python
a = 10 # Global Scope.

def function():
    b = 200 # Local Scope
    print(a)
    
function()


print(b) # It will look only on global scope. 

# Why error!!!
```

```python
# If you want to acces the local scope as global scope you can use the global function 
# OR 
# If you want to change the global scope you can use this 

b = 39

def function():
    global b # We are changed the global scope (it is a keyword help to access the global variable)
    b = 20
    print(b)
    
    
function()
    
print(b)
```

```python
a = 30

def function(): 
    global b   # If There is no global variable global keyword make a new varaible b and assign a given value
    b =20 
    print(b)
    
function()

print(b)
```

### Passing a List in Function

```python
lst = [12,23,42,54,64,2,1]


def identifier(lst):
    even = 0
    odd = 0
    for i in lst:
        if i%2==0:
            even = even + 1
        else:
            odd = odd + 1
        
    return even, odd
    
    
    
even, add = identifier(lst)
```

## Filter 

* Filter elements filter the elements(iterabels) based on the expressions. 
* Filter the unwanted elements. 

#### Synatx - filter(function, iterabels) 
##### It will return True or False

* It will return the iterator if you want to see the result use list or tuple

```python
for x in range(1,11):
    if x %2==0:
        print(x)
        
#____________________________USING FILTER___________________________________

list(filter(lambda x: x%2 ==0, range(1,11)))
```

```python
a = [1,2,3,7]
list(filter(lambda a:a>5, a))
```

```python
a = [1,2,3]
b = [1,2,3]

list(map(lambda x,y: x+y, a,b))
```

```python
list(filter(lambda x,y: x+y, a,b)) # Because filter accept only one iterable
```

### Reduce Function


* Reduce the iterable to single element using some functions. 

Uses - If you want to find out the sum and product of all the list you can use this funciton. 


#### Syntax- reduce(function, iterables)

```python
import functools # Function tools 


functools.reduce(lambda x,y: x+y, a) # finding sum of one list 
```

```python
functools.reduce(lambda x,y: x+y, a,b) # Because both reduce and filter function accept only one iterables
```

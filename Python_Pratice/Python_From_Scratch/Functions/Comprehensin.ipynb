# List Comprehension

### Basic Syntax: 

* Syntax : [ **expression** for **item** in **interable** ]
* Syntax : [ **x** for **x** in **range(100)** ]

### Condition Syntax: 

* Syntax: [ **expression** for **item** in **iterable** if **conditional** ] ( **Single IF**)
* Syntax: [ **expresssion** if **conditional** else **statement** for **item** in **iterable**] (**Both IF and ELSE**)

```python
ls = []
for i in range(1,101):
    if i %2!=0:
        ls.append(i)
        
#________________________________________________List Comprehension____________________________________________________________

# Basic 
l = [x for x in range(1,100)]

# Condition Syntax (Single IF)
l = [ x for x in range(1,100) if x %2==0]

# Condition Syntax (Both If and else)
l = [ x if x>2 else x+1 for x in range(1,11)]


lb = []
l = [i if i%2==0 else lb.append(i) for i in range(101)]
```

### Examples: 

```python
# Multiply the list with 10.. 

num = [1,2,3,4]

ls = [i*10 for i in num ]
print(ls)
```

```python
# changing the case to upper..

word = ['string', 'a', 'aravind']

ls = [i.upper() for i in word]

print(ls)
```

```python
# Extract numbers from list 

word = 'aravind22052001'

ls = [i for i in word if i.isdigit()]

print(ls)
```

```python
# Nested list comprehension (retrive all the first digit from the nested list)

ls = [[1,2], [3,4], ['a','b']]

lse = [i[0] for i in ls]
print(lse)
```

```python
# list comprehesion on function.. 

def square(x):
    return x*x

ls = [square(x) for x in range(1,11)]
print(ls)
```

```python
# To add two list 

lis1 = [1,2,3]
list2 = [1,2,3]

ls = [ x+y for x in lis1 for y in list2]

# adding 2 list in same position 
ls = [lis1[i] + list2[i] for i in range(0,len(lis1))]
ls
```

### List Comprehension is Perfect substitute for MAP(), FILTER(), REDUCE() Function

#### List With Lambda Function

```python
num = [1.2, 2.3, 4.5]

# Using int() Function 
ls = list(map(int, num))
print(ls)


# Using lambda function 

ls = list(map(lambda x: int(x), num))
print(ls)

# __________________________Time to List Comprehension______________________________

ls = [int(i) for i in num]
print(ls)
```

```python
import timeit

timeit.timeit('list(map(lambda x: int(x), [1.2, 2.3, 4.5]))')
```

```python
timeit.timeit('[int(i) for i in [1.2, 2.3, 4.5]]')
```

* This List comprehension is fast compare to both the things

### List with Reduce Function 

```python
import functools 

ls = functools.reduce(lambda x,y: x+y, [1,2,3,4])
print(ls)

# List Comprehension 

ls = sum([i for i in [1,2,3,4]])
print(ls)
```

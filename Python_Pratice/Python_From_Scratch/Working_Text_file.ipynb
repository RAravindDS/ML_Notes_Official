# Opeining Text file

Syntax::  **open('filename','filemode')**

```python
file = open('testing.txt', 'r') # Default 2nd parameter 'r' - read
```

### Tell 

* Tell helps to tell the index of the text file string

```python
file.tell() # it tells the index position. 
```

```python
# For reaading we use 'read()'
file.read()
```

```python
file.read() # we will get empty string. Because it changes the cursor position so it shows after the words,
# so we will get the empty string
```

```python
file.tell() # It tells index.. 
```

### Seek 

* If you want to repeat the reading you can use seek method
Syntax:: seek('index')

```python
file.seek(0) # it changes the cursor positon, so we can read multiple times by using this. 
```

```python
file.read()
```

```python
file.close()

# We need to close the file, "it is mandatory"
```
